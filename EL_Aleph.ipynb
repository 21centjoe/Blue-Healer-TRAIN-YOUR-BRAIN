```python
import random

def roll_dice_with_color_coded_die(num_dice=11):
    """
    Simulates rolling a specified number of standard dice (1-6) plus one color-coded die.

    Args:
        num_dice (int): The number of standard 6-sided dice to roll.

    Returns:
        tuple: A tuple containing:
               - list: The results of the standard dice rolls.
               - dict: The result of the color-coded die, including its value and color.
    """
    # Roll the standard dice
    standard_rolls = [random.randint(1, 6) for _ in range(num_dice)]

    # Roll the color-coded die
    possible_colors = ['red', 'blue', 'green', 'yellow', 'purple'] # Expanded color options
    color_coded_value = random.randint(1, 6)
    color_coded_color = random.choice(possible_colors)

    color_coded_die_result = {
        'value': color_coded_value,
        'color': color_coded_color
    }

    return standard_rolls, color_coded_die_result

# Example usage:
standard_rolls, color_die = roll_dice_with_color_coded_die()

print(f"Standard dice rolls ({len(standard_rolls)} dice): {standard_rolls}")
print(f"Color-coded die roll: Value = {color_die['value']}, Color = {color_die['color']}")
```

```python
def determine_superpower(color_die_result):
    """
    Determines a superpower and its condition based on the color and value of the color-coded die.

    Args:
        color_die_result (dict): A dictionary containing 'value' (int) and 'color' (str) of the color-coded die.

    Returns:
        tuple: A tuple containing the superpower (str) and its condition (str).
    """
    color = color_die_result['color']
    value = color_die_result['value']

    superpower = ""
    condition = ""

    if color == 'red':
        superpower = "Super Strength"
        if value <= 2:
            condition = "Power drains quickly after use."
        elif value <= 4:
            condition = "Power weakens significantly in cold environments."
        else:
            condition = "Requires intense focus; easily distracted."
    elif color == 'blue':
        superpower = "Telekinesis"
        if value <= 2:
            condition = "Limited to small, non-living objects."
        elif value <= 4:
            condition = "Causes extreme mental fatigue after prolonged use."
        else:
            condition = "Ineffective when surrounded by loud noises."
    elif color == 'green':
        superpower = "Nature Control"
        if value <= 2:
            condition = "Only works on decaying plant life."
        elif value <= 4:
            condition = "Must be barefoot on natural ground to activate."
        else:
            condition = "Effect is reversed if exposed to artificial light for too long."
    elif color == 'yellow':
        superpower = "Lullaby Singer (Frog)"
        if value <= 2:
            condition = "Only affects snakes, cats, and dogs; requires absolute silence to work."
        elif value <= 4:
            condition = "Effective only at night or in dimly lit areas; drains energy quickly."
        else:
            condition = "Can make targets sleep, but prolonged use causes the frog to croak uncontrollably for an hour."
    elif color == 'purple':
        superpower = "Invisibility"
        if value <= 2:
            condition = "Only affects clothing; body remains visible."
        elif value <= 4:
            condition = "Momentary invisibility only, with a long cooldown."
        else:
            condition = "Superpower stops immediately if you get wet."
    else:
        superpower = "No Superpower"
        condition = "No special conditions."

    return superpower, condition

# Example usage with the previously rolled color_die:
superpower, condition = determine_superpower(color_die)

print(f"Based on your die roll (Color: {color_die['color']}, Value: {color_die['value']}):")
print(f"Superpower: {superpower}")
print(f"Condition: {condition}")
```

```python
def calculate_cornhole_score(washer_landings):
    """
    Calculates the total score for a round of magnetized washer cornhole.

    Args:
        washer_landings (list): A list of strings indicating where each washer landed.
                                Possible values: 'hole', 'board', 'off_board'.

    Returns:
        int: The total score for the round.
    """
    score = 0
    for landing in washer_landings:
        if landing == 'hole':
            score += 5  # 5 points for landing in the hole
        elif landing == 'board':
            score += 1  # 1 point for landing on the board (due to magnetism, not in hole)
        # 'off_board' or any other value gets 0 points
    return score

# Example usage:
player_washers_round1 = ['hole', 'board', 'hole', 'off_board', 'board']
player_score = calculate_cornhole_score(player_washers_round1)

print(f"Player's washer landings: {player_washers_round1}")
print(f"Total score for the round: {player_score} points")

# Another example:
player_washers_round2 = ['board', 'board', 'board', 'off_board', 'off_board']
player_score_2 = calculate_cornhole_score(player_washers_round2)
print(f"\nPlayer's washer landings: {player_washers_round2}")
print(f"Total score for the round: {player_score_2} points")
```

### Leaderboard Class

This class will manage and display scores for different players across various mini-games.

```python
class Leaderboard:
    def __init__(self):
        self.scores = {}

    def add_score(self, player_name, game_name, score):
        if player_name not in self.scores:
            self.scores[player_name] = {
                'cornhole': 0,
                'fruit_balancing': 0,
                'surf_boy': 0
            }
        # Update score for the specific game
        if game_name in self.scores[player_name]:
            self.scores[player_name][game_name] = score
        else:
            # If a new game type is added later, it will be handled
            print(f"Warning: Game '{game_name}' not pre-defined. Adding score for {player_name} in {game_name}.")
            self.scores[player_name][game_name] = score

    def get_total_score(self, player_name):
        """Calculates the total score for a given player."""
        if player_name in self.scores:
            return sum(self.scores[player_name].values())
        return 0

    def display_leaderboard(self):
        """
        Displays the leaderboard, sorted by total score.
        """
        print("\n--- Game Leaderboard ---")
        if not self.scores:
            print("No scores recorded yet.")
            return

        leaderboard_data = []
        for player, game_scores in self.scores.items():
            total = self.get_total_score(player)
            leaderboard_data.append({'player': player, 'total_score': total, **game_scores})

        # Sort by total score in descending order
        leaderboard_data.sort(key=lambda x: x['total_score'], reverse=True)

        # Print header dynamically based on available games
        game_names = ['cornhole', 'fruit_balancing', 'surf_boy'] # Order for display
        header_parts = ["Player ", "Total "] + [game.replace('_', ' ').title().ljust(15) for game in game_names]
        header_format = "{:<15} " * (len(game_names) + 2)
        print(header_format.format("Player", "Total", *[game.replace('_', ' ').title() for game in game_names]))
        print("-" * (15 * (len(game_names) + 2)))

        # Print scores
        for entry in leaderboard_data:
            row_values = [entry['player'], entry['total_score']]
            for game in game_names:
                row_values.append(entry.get(game, 0)) # Use .get to handle cases where a game score might be missing
            print(header_format.format(*row_values))
        print("------------------------\n")

# Initialize the leaderboard
game_leaderboard = Leaderboard()
```

### Simulate Cornhole Game (Regenerated)

The `simulate_cornhole_game` function is re-added to allow for simulation and score tracking.

```python
import random

def simulate_cornhole_game(player_name, num_rounds=10, washers_per_round=5):
    """
    Simulates multiple rounds of magnetized washer cornhole and calculates total score for a player.

    Args:
        player_name (str): The name of the player.
        num_rounds (int): The number of rounds to simulate.
        washers_per_round (int): The number of washers thrown per round.

    Returns:
        int: The total accumulated score across all rounds for the player.
    """
    total_game_score = 0
    possible_landings = ['hole', 'board', 'off_board']

    print(f"\nSimulating {num_rounds} rounds of Cornhole for {player_name} (with {washers_per_round} washers per round):\n")

    for i in range(1, num_rounds + 1):
        # Simulate landings for the current round
        # Weights: 30% hole, 40% board, 30% off_board
        current_round_landings = random.choices(possible_landings, weights=[0.3, 0.4, 0.3], k=washers_per_round)
        round_score = calculate_cornhole_score(current_round_landings)
        total_game_score += round_score

        print(f"Round {i}: Landings = {current_round_landings}, Score = {round_score} points")

    print(f"\n--- Cornhole Game Over for {player_name} ---")
    print(f"Total accumulated score for {player_name}: {total_game_score} points")
    return total_game_score

# Run the simulation for Player 1 and add to leaderboard
player1_cornhole_score = simulate_cornhole_game(player_name="Player 1", num_rounds=10, washers_per_round=5)
game_leaderboard.add_score("Player 1", "cornhole", player1_cornhole_score)

# Run the simulation for Player 2 and add to leaderboard
player2_cornhole_score = simulate_cornhole_game(player_name="Player 2", num_rounds=10, washers_per_round=5)
game_leaderboard.add_score("Player 2", "cornhole", player2_cornhole_score)
```
